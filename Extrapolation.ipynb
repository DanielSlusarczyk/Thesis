```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import xgboost as xgb

def generate_noisy_linear_data(n_points=100, slope=2, intercept=1, noise_std=1, a=0, b=10):
    # Generowanie liniowych danych
    x = np.linspace(a, b, n_points)
    y = slope * x + intercept

    # Dodawanie szumu do danych
    noise = np.random.normal(0, noise_std, n_points)
    y_noisy = y + noise

    # Tworzenie DataFrame
    data = {'x': x, 'y': y_noisy}
    df = pd.DataFrame(data)

    return df

# Przykładowe użycie
df_learn = generate_noisy_linear_data()
df_test = generate_noisy_linear_data(a=-2, b=12)

fig=go.Figure()

fig.add_trace(go.Scatter(
    x = df_learn['x'],
    y = df_learn['y'],
    name = 'Dane uczące',
    mode = 'markers'
))

fig.add_trace(go.Scatter(
    x = df_test['x'],
    y = df_test['y'],
    name = 'Dane uczące',
    mode = 'markers'
))

fig.update_layout(title='Problem ekstrapolacji danych', height=400, width=1000)

fig.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = xgb.XGBRegressor (n_estimators = 10)
clf.fit(df_learn['x'], df_learn['y'])
xgb_predictions = clf.predict(df_test['x'])

clf = RandomForestRegressor(n_estimators=10)

clf.fit(df_learn[['x']], df_learn[['y']])
rf_predictions = clf.predict(df_test[['x']])

fig=go.Figure()

fig.add_trace(go.Scatter(
    x = df_learn['x'],
    y = df_learn['y'],
    name = 'Dane uczące',
    mode = 'markers',
    marker=dict(size=5)
))

fig.add_trace(go.Scatter(
    x = df_test['x'],
    y = df_test['y'],
    name = 'Dane testowe',
    mode = 'markers',
    marker=dict(size=5)
))

fig.add_trace(go.Line(
    x = df_test['x'],
    y = xgb_predictions,
    name = 'XGBoost',
))

fig.add_trace(go.Line(
    x = df_test['x'],
    y = rf_predictions,
    name = 'Random Forest',
))

fig.update_layout(title='Problem ekstrapolacji danych', height=500, width=1000, xaxis_title="X", yaxis_title="Y")

fig.show()
```
